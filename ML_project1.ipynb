```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#import dataset
df=pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
x = df.drop('Class', axis=1)
y = df['Class']
```

```python
x.head()
```

```python
y.head()
```

##Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.3, random_state=42,stratify=y)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

##Normmalising the amount column

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
y_train = scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test = scaler.transform(y_test.values.reshape(-1, 1))
```

#Applying SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
```

```python
x_train,y_train = smote.fit_resample(x_train,y_train)
```

```python
import pandas as pd
pd.Series(y_train.flatten()).value_counts()
```

```python
x_train.shape,y_train.shape
```

##Training the models

####MODEL 1  Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#model prediction
pred = model.predict(x_test)
```

##MODEL EVALUATION

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
```

```python
print("Accuracy:",accuracy_score(y_test,pred),"\n","Confusion Matrix:\n",confusion_matrix(y_test,pred),"\n","\n",classification_report(y_test,pred),"\n","ROC AUC Score:",roc_auc_score(y_test,pred))
```

###MODEL 2 RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2 = RandomForestClassifier(n_jobs=-1,n_estimators=50,max_depth=5)
```

```python
model2.fit(x_train,y_train)
```

```python
pred2 = model2.predict(x_test)
```

##model2 evaluation

```python
print("Accuracy:",accuracy_score(y_test,pred2),"\n","Confusion Matrix:\n",confusion_matrix(y_test,pred2),"\n","\n",classification_report(y_test,pred2),"\n","ROC AUC Score:",roc_auc_score(y_test,pred2))
```

##MODEL3 XGBoost

```python
from xgboost import XGBClassifier
```

```python
scale_pos_weight = len(y_train[y_train == 0]) / len(y_train[y_train == 1])
```

```python

model3 = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_pos_weight,
    n_jobs=-1,
    random_state=42,
    eval_metric='auc'
)
```

```python
model3.fit(x_train,y_train)
```

```python
pred3 = model3.predict(x_test)
```

##model3 evaluation

```python
print("Accuracy:",accuracy_score(y_test,pred3),"\n","Confusion Matrix:\n",confusion_matrix(y_test,pred3),"\n","\n",classification_report(y_test,pred3),"\n","ROC AUC Score:",roc_auc_score(y_test,pred3))
```
